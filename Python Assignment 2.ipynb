#### l. Plot the skewness - All Numeric Variables

```python
x1 = data['UnitPrice']
y1 =  1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  )

plt.plot(x1,y1)
print('Sknewness of Quantity is: ', data['UnitPrice'].skew())
```

```python
x1 = data['CustomerID']
y1 =  1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  )

plt.plot(x1,y1)
print('Sknewness of Quantity is: ', data['CustomerID'].skew())
```

## 2. Missing values Treatment

```python
data.isna().sum()
```

In the above boxplot for CustomerID we saw no outliers hence we will fill the missing values with the mean 


```python
m = round(data['CustomerID'].mean())
m
```

```python
data['CustomerID'].fillna(m,inplace=True)
```

```python
data['CustomerID'].isna().sum()
```

Filling categorical data with the mode 

```python
m = str(data['Description'].mode())
print(m)
```

```python
data['Description'].replace(np.nan,m, inplace=True)
```

```python
data['Description'].isna().sum()
```

## 3. Remove duplicate rows

```python
data.drop_duplicates()
```

## 4. Remove rows which have negative values in Quantity column 


```python
data.drop(data[data["Quantity"]<0].index,inplace=True)
```

```python
data.shape[0]
```

## 5. Add the columns - Month, Day and Hour for the invoice 

```python
data['InvoiceDate'] = pd.to_datetime(data.InvoiceDate, format='%m/%d/%Y %H:%M')
```

```python
data.head(1)
```

```python
data.insert(loc=5, column='Month', value= data.InvoiceDate.dt.month)
data.insert(loc=6, column='Day', value=data['InvoiceDate'].dt.day)
data.insert(loc=7, column='Hour', value=data['InvoiceDate'].dt.hour)
```

```python
data.head(1)
```

#### 6. How many orders made by the customers? 


```python
print("Total orders placed by the customers: ", data['Quantity'].value_counts().sum())
```

#### 7. TOP 5 customers with higher number of orders 

```python
data.groupby(by='CustomerID', as_index=False)['InvoiceNo'].count().sort_values(by='InvoiceNo', ascending=False).head()
```

#### 8. How much money spent by the customers? 

```python
#adding column am
data['Amt'] = data['Quantity']*data['UnitPrice']
```

```python
print("Total orders placed by the customers: $", round(data['Amt'].sum(),2))
```

#### 9. TOP 5 customers with highest money spent 

```python
top_amt = data.groupby(by='CustomerID', as_index=False)['Amt'].sum()
```

```python
top_amt.sort_values(by='Amt', ascending=False).head()
```

#### 10. How many orders per month? 


```python
orders_month = data.groupby('InvoiceNo')['Month'].unique().value_counts().sort_index().plot(kind='bar',figsize=(20,6))
orders_month.xlabel=('Month')
orders_month.ylabel=('Count of Orders')
orders_month.set_xticklabels(('Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec'),fontsize=15)
plt.show()
```

#### 11. How many orders per day?

```python
data.insert(loc=12,column='DayOfWeek',value=(data['InvoiceDate'].dt.dayofweek)+1) 
```

```python
data.shape
```

```python
order_day = data.groupby('InvoiceNo')['DayOfWeek'].unique().value_counts().sort_index().plot(kind='bar',figsize=(20,6))
order_day.xlabel=('Day of the Week')
order_day.ylabel=('Count of Orders')
order_day.set_xticklabels(('Mon','Tue','Wed','Thurs','Fri','Sun'),fontsize=15)
plt.show()
```

#### 12. How many orders per hour? 

```python
order_hr = data.groupby(by='InvoiceNo')['Hour'].unique().value_counts().plot(kind='bar',figsize=(15,6))
order_hr.xlabel=('Day of the Week')
order_hr.ylabel=('Count of Orders')
order_hr.set_xticklabels(range(6,22), rotation='horizontal', fontsize=15)
plt.show()
```

#### 13. How many orders for each country?
